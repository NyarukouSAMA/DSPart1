```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
masses = [0.055, 0.815, 1.0, 0.107, 317.8, 95.0, 14.6, 17.147]
```

```python
plt.plot(planets, masses)
plt.show()
```

```python
masses[planets.index('Earth')] / masses[planets.index('Mercury')]
```

```python
plt.plot(planets, masses)
plt.yscale(value = 'log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planets)
planet_info
```

```python
planet_info.plot()
plt.show()
```

```python
planet_info.plot(kind = 'bar')
plt.show()
```

```python
planet_info.plot(kind = 'bar', logy = True)
plt.show()
```

```python
planet_info.plot(kind = 'barh', logx = True)
plt.show()
```

```python
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 6,4.5
```

```python
x = np.linspace(-5, 5, 101)
x
```

```python
x2 = x ** 2
x2
```

```python
plt.plot(x, x2)
plt.show()
```

```python
plt.plot(x, x2)
plt.title(' f(x)')
plt.show()
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(x, x2)
plt.title(' f(x)')
plt.show()
```

```python
plt.plot(x, x2)
plt.title(' f(x)', fontsize = 16, fontweight = 'bold', color = 'DarkRed')
plt.show()
```

```python
plt.plot(x, x2)
plt.title(' f(x)', fontsize = 16, fontweight = 'bold', color = '#808080')
plt.show()
```

```python
title_dict = {
    'fontsize': 16,
    'fontweight': 'bold',
    'color': '#808080',
    'family': 'Calibri'
}

plt.plot(x, x2)
plt.title(' f(x)', fontdict = title_dict)
plt.show()
```

```python
plt.plot(x, x2)
plt.title(' f(x)', fontdict = title_dict, loc = 'left')
plt.show()
```

```python
plt.plot(x, x2)
plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

plt.show()
```

```python
x3 = x ** 3
x3
```

```python
plt.plot(x, x2)
plt.plot(x, x3)

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2)
plt.plot(x, x3)

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

plt.legend(labels = ['x2', 'x3'])

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label = 'x ** 2')
plt.plot(x, x3, label = 'x ** 3')

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

plt.legend()

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label = 'x ** 2')
plt.plot(x, x3, label = 'x ** 3')

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

plt.legend(loc = 'lower right', fontsize = 8)

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label = 'x ** 2')
plt.plot(x, x3, label = 'x ** 3')

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

legend = plt.legend(loc = 'lower right', fontsize = 8, frameon = False)
plt.setp(legend.get_texts(), color = 'grey')

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label = 'x ** 2', color = '#3bb08f')
plt.plot(x, x3, label = 'x ** 3', linestyle = 'dotted')

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

legend = plt.legend(loc = 'lower right', fontsize = 8, frameon = False)
plt.setp(legend.get_texts(), color = 'grey')

plt.grid()
plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label = 'x ** 2', color = '#3bb08f')
plt.plot(x, x3, label = 'x ** 3', linestyle = 'dotted')

plt.title(' f(x)', fontdict = title_dict)

plt.xlabel(' x')
plt.ylabel(' f(x)')

legend = plt.legend(loc = 'lower right', fontsize = 8, frameon = False)
plt.setp(legend.get_texts(), color = 'grey')

plt.grid(axis = 'y', color = 'lightgrey')
plt.show()
```
