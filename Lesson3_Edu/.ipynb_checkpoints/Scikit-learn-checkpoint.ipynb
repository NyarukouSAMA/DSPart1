## Линейная регрессия

Линейная регрессия - алгоритм, предназначенный для предсказания непрерывной величины,
например, динамики роста цен на недвижимость. Задача предсказани таких величин - это задача регрессии.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
data = boston.data
```

```python
data.shape
```

```python
data
```

Target - величина, которую нужно предсказать для датасета.

```python
target = boston.target
```

feature_name - названия признаков.

```python
feature_names = boston.feature_names
```

```python
feature_names
```

Описание датасета:

```python
for line in boston.DESCR.split('\n'):
    print(line)
```

#### Подгоовка данных

Тут мы готовим данные для модели линейной регрессии:
- X - наборы признаков, по которым идет оценка предиктивной величины
- Y - набор значений предиктивной величины

```python
X = pd.DataFrame(data, columns = feature_names)
```

```python
X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=['price'])
y.head()
```

```python
y.info()
```

После подготовки наборов признаков мы разбиваем наборы на тренировачный и тестовый:
- Тренировачный - набор, при помощи которого обучают модель.
- Тестовый - набор, на котором модель тестируют

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

#### Построение модели линейной регрессии

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

Задача линейной регрессии - поиск подходящих коэффициентов схождения признаков (**w**) при признаках x для вычисления целевой переменной y, минимизируя ошибку e:

### $$y = w_{0} + w_{1}*x_{1} + w_{2}*x_{2} + ... + w_{m}*x_{m} + e$$

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    'y_test' : y_test['price'],
    'y_pred' : y_pred.flatten()
});
check_test.head(10)
```

### Метрики оценки качества регрессионной модели

Из сравнения тестовой выборки и предсказания можно заметить, что часть оценок из предсказательной модели отличаются от реальных тестовых данных. Для того, что бы понять, насколько точна полученная предсказательная модель, мы должны определить специальные показатели модели (метрики оценки качества регрессионных моделей).

#### Средняя квадратичная ошибка

Средняя квадратичная ошибка - это средний квадрат разности между предсказанной и реальной величиной

```python
check_test['error'] = check_test.y_pred - check_test.y_test
```

```python
check_test.head()
```

```python
initial_mse = (check_test.error ** 2).mean()
initial_mse
```

```python
from sklearn.metrics import mean_squared_error
```

```python
initial_mse = mean_squared_error(y_test, y_pred)
initial_mse
```

#### Средняя абсолютная ошибка

Средняя абсолютная ошибка - это средний модуль разности между предсказанной и реальной величиной

```python
(np.abs(check_test.error)).mean()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

#### Метрика $R^2$

$R^2$ - это коэффициент детерминации, который характеризует долю вариации (дисперсии) результативного признака y, объясняемую регрессией, в общей вариации (дисперсии) y. Регрессионное уравнение считается приемлемым, если $R^2 > 0.5$. При значении $R^2 >= 80\%$ - регрессионная модель считается хорошей. Единица - означает функциональную линейную зависимость.

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

#### Просмотр коэфициентов линейной регрессии

$w_0$

```python
lr.intercept_
```

 $w_1...w_m$

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.barh(X_train.columns, lr.coef_.flatten())
plt.xlabel('Вес признака')
plt.ylabel('Признак')
```

#### Стандартизация признаков

По данной диаграмме видно, что вклад каждого признака в итоговую оценку оценить сложно, поэтому признаки следует масштабировать.
Средством масштабирования признаков в выборке является стандартизация. Для стандартизации признака нужно от значения признака отнять среднее и поделить на среднеквадратичное отклонение:
### $$X_{scaled} = x-x_{mean} \over \sigma_x$$

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
lr.fit(X_train_scaled, y_train)
```

```python
plt.barh(X_train.columns, lr.coef_.flatten())
plt.xlabel('Вес признака')
plt.ylabel('Признак')
```
